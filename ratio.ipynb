```python
df=pd.merge(table2,table1[['emp_id','dept']], on='emp_id')
mnth_att=df.groupby(['dept', 'month']).agg(
    total_employees=('emp_id', 'count'),
    total_present=('attendies', 'sum')
).reset_index()
```

```python
merged_df = pd.merge(table2, table1, on='emp_id')
attendance_by_month = merged_df.groupby(['dept', 'month']).agg(
    total_attendance=('attendies', 'sum'),
    total_employees=('emp_id', 'count')
).reset_index()
jan_attendance = attendance_by_month[attendance_by_month['month'] == 'Jan-24'][['dept', 'total_employees']]
attendance_by_month = pd.merge(attendance_by_month, jan_attendance, on='dept', suffixes=('', '_jan'))
attendance_by_month['attendance_ratio'] = attendance_by_month['total_attendance'] / attendance_by_month['total_employees_jan']
attendance_by_month['attendance_ratio'].fillna(1, inplace=True)
print(attendance_by_month[['dept', 'month', 'attendance_ratio']])
```
