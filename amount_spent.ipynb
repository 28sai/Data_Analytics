```python
df['year']=df['date'].dt.year
df['month']=df['date'].dt.month
min_year=df['year'].min()
max_year=df['year'].max()
ans=df.groupby(['name', 'year', 'month'])['amt'].sum()
d={}
for (i,yr,mn), amt in ans.items():
    if i not in d:
        d[i]=[0]*((max_year-min_year+1)*12) 
    index=(yr-min_year)*12+(mn-1)
    d[i][index]=amt
for person, years_data in d.items():
    print(f"Data for {person}:")
    for year in range(min_year, max_year + 1):
        for month in range(1, 13):
            amt = years_data.get((year, month), 0)
            print(f"Year {year} - Month {month}: {amt}")
        
    print("-" * 50)
```

```python
import pandas as pd
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
ans = df.groupby(['name', 'year', 'month'])['amt'].sum()
d = {}
for (i, year, month), amt in ans.items():
    if i not in d:
        d[i] = {}
    d[i][(year, month)] = amt
for person, years_data in d.items():
    print(f"Data for {person}:")
    for year in sorted(set(df['year'])):
        for month in range(1, 13):
            amt = years_data.get((year, month), 0)
            print(f"Year {year} - Month {month}: {amt}")
    print("-" * 50)
```
